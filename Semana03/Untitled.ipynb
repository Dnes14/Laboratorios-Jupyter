```python
# Importamos las librerias a utilizar

import pandas as pd
import numpy as np
```

## Parte A

```python
#Leemos los datos

data = pd.read_csv("banca.csv", sep = ';')
data
```

```python
# Seleccionamos la partes numericas

v_numericas = data.select_dtypes(["float64","int64"])

```

```python
#convirtiendo el dataframe a matriz

X = v_numericas.to_numpy()
X
```

```python
X.shape
```

## PARTE B

```python
# Vector de medias para cada variables

Medias = np.mean(X,axis = 0)
Medias
```

```python
# Resta de la matriz de datos y el vector de medias

Resta= X - Medias
Resta
```

```python
# Matriz de covariana de las variables
Covarianza = np.cov(X, rowvar=False)
Covarianza
```

```python
# Inversa de la matriz  de covarianza de las variables

Inversa = np.linalg.inv(Covarianza)
Inversa
```

## Parte C

```python
# Multiplicando la 
n = X.shape[0]
valores = []
for i in range (n):
    valor = np.dot(np.dot(Resta[i],Inversa),Resta[i].T)
    valores.append(valor)
valores
```

```python
from scipy.spatial.distance import mahalanobis 

f = X.shape[0]
valoresmaha = []
for i in range (f):
    valor = mahalanobis(X[i],Medias,Inversa)**2
    valoresmaha.append(valor)
valoresmaha
```

```python
data['Mahalanobis'] = valores
data ['Indices'] = data.index
data
```

```python
data = data[data["Mahalanobis"] > 100 ]
data = data.drop(['Mahalanobis', 'Indices'], axis =1)
data
```

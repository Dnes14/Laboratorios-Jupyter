```python
# Coordenadas de la letra I
I = np.array([(0,0,1), (3,0,1), (3,1,1), (2,1,1), (2,4,1), (3,4,1), (3,5,1), 
              (0,5,1), (0,4,1), (1,4,1), (1,1,1), (0,1,1), (0,0,1)])
I
```

## B: Aplique una transformaciÃ³n de escala con s = 3 y grafique la imagen original junto con su versiÃ³n transformada.

```python
# Matriz de cambio de escala
s=3 #Escalar
S=np.array ([[s, 0., 0.], [0., s, 0.] ,[0. , 0., 1.]])
S
```

```python
# Aplicar la transformaciÃ³n

Ix = [] 
Iy = [] 
for row in I: 
    output_rown = (S.dot(row)) 
    x, y, z = output_rown 
    Ix.append(x) 
    Iy.append(y) 
T = [Ix, Iy] 
```

```python
#Graficamos

ax= plt.gca() 
plt.plot(I[:, 0], I[:, 1],color="blue", label="Original") 
plt.plot(T[0], T[1], color="red", label="Transformado") 
ax.set_xticks (np.arange(-3,12,1)) 
ax.set_yticks (np.arange(-3,23,1)) 
ax.set_aspect("equal", adjustable="box") 
plt.grid() 
plt.rcParams["figure.figsize"] = (8,4) 
plt.legend(loc="upper center") 
plt.show()
```

## C: Aplique una transformaciÃ³n de deformaciÃ³n horizontal/vertical con h = â€“4 y v = 3, y grafique la imagen original junto con su versiÃ³n transformada.

```python
# Matriz de deformaciÃ³n horizontal/vertical
h = -4
v =  3
D = np.array([[1., h, 0.], [v, 1., 0.], [0., 0., 1.]])
D
```

```python
# Aplicar la transformaciÃ³n
Ix1 = [] 
Iy1 = [] 
for row in I: 
    output_rown = D.dot(row)
    x, y, z = output_rown
    Ix1.append(x) 
    Iy1.append(y) 
T2 = [Ix1, Iy1] 
```

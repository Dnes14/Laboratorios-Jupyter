```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Coordenadas de la figura
I = np.array([(0,0,1), (3,0,1), (3,1,1), (2,1,1), (2,4,1), (3,4,1), (3,5,1), 
              (0,5,1), (0,4,1), (1,4,1), (1,1,1), (0,1,1), (0,0,1)])
I
```

### Transformacin de Escala con s=3

```python
# Definir la matriz de escala
s = 3
S = np.array([[s, 0, 0], 
              [0, s, 0], 
              [0, 0, 1]])
```

```python
Ix = []
Iy = []
for row in I:
    output_row = S.dot(row)
    x, y, z = output_row
    Ix.append(x)
    Iy.append(y)
T = [Ix, Iy]
T
```

```python
plt.figure(figsize=(8, 4))
ax = plt.gca()
plt.plot(I[:, 0], I[:, 1], color="blue", label="Original")
plt.plot(T[0], T[1], color="red", label="Escala s=3")
ax.set_xticks(np.arange(-2, 12, 1))  
ax.set_yticks(np.arange(-2, 22, 1))  
ax.set_aspect("equal", adjustable="box")
plt.grid()
plt.legend(loc="upper center")
plt.show()
```

## Transformacin de Deformacin Horizontal/Vertical con h=4 y v=3

```python
# Definir la matriz de deformacin
h, v = -4, 3
D = np.array([[1, h, 0], 
              [v, 1, 0], 
              [0, 0, 1]])
```

```python
Ix = []
Iy = []
for row in I:
    output_row = D.dot(row)
    x, y, z = output_row
    Ix.append(x)
    Iy.append(y)
T = [Ix, Iy]
T
```

```python
# Graficar la figura original y la transformada
plt.figure(figsize=(8, 4))
ax = plt.gca()
plt.plot(I[:, 0], I[:, 1], color="blue", label="Original")
plt.plot(T[0], T[1], color="red", label="Deformacion h, v = 4, 3")
ax.set_xticks(np.arange(-21, 12, 1))  
ax.set_yticks(np.arange(-2, 22, 1))  
ax.set_aspect("equal", adjustable="box")
plt.grid()
plt.legend(loc="upper center")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from IPython.display import Image
Image("letra I.png")
```

```python
I = np.array([(0,0,1), (3,0,1), (3,1,1), (2,1,1), (2,4,1), (3,4,1), (3,5,1), 
              (0,5,1), (0,4,1), (1,4,1), (1,1,1), (0,1,1), (0,0,1)])
I
```

```python
s=3 
S=np.array ([[s, 0, 0.], [0., s, 0.] ,[0. , 0., 1.]])
```

```python
Ix = [] 
Iy = [] 
for row in I: 
    output_rown = (S.dot(row)) 
    x, y, z = output_rown 
    Ix.append(x) 
    Iy.append(y) 
T = [Ix, Iy] 
T
```

```python
ax= plt.gca() 
plt.plot(I[:, 0], I[:, 1],color="blue", label="Original") 
plt.plot(T[0], T[1], color="red", label="Transformado") 
ax.set_xticks (np.arange(-3,12,1)) 
ax.set_yticks (np.arange(-3,23,1)) 
ax.set_aspect("equal", adjustable="box") 
plt.grid() 
plt.rcParams["figure.figsize"] = (8,4) 
plt.legend(loc="upper center") 
plt.show()
```

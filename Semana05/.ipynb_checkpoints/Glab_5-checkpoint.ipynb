```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("diabetes.csv")
data
```

```python
data1 = data.drop("Outcome", axis = 1)
data1
```

```python
data_std = StandardScaler().fit_transform(data1)
A = np.cov(data_std, rowvar = False)
A
```

```python
M = np.linalg.eig(A)
Autovalores, Autovectores = M[0], M[1]
print("Autovalores: \n",Autovalores)
print("Autovectores: \n",Autovectores)
```

```python
from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
```

```python
# Entrenando y escalando los datos
pca_pipe = make_pipeline(StandardScaler(),PCA())
pca_pipe.fit(data1)
modelo_pca = pca_pipe.named_steps['pca']
```

```python
# Convirtiendo el array a dataframe
df1 = pd.DataFrame(data=modelo_pca.explained_variance_,columns=['Autovalor'],
                  index=["Î»1","Î»2","Î»3","Î»4","Î»5","Î»6","Î»7","Î»8"])
df2 = pd.DataFrame(data=modelo_pca.explained_variance_ratio_*100,columns=['%Varianza'],
                  index=["Î»1","Î»2","Î»3","Î»4","Î»5","Î»6","Î»7","Î»8"])
df3 = pd.DataFrame(data=modelo_pca.explained_variance_ratio_.cumsum()*100,columns=['%Varianza acum.'],
                  index=["Î»1","Î»2","Î»3","Î»4","Î»5","Î»6","Î»7","Î»8"])
pd.concat([df1,df2,df3],axis=1)
```

```python
# convirtiendo el array a dataframe
data2 = pd.DataFrame(data= modelo_pca.components_,columns=data1.columns,
                    index = ['CP1','CP2','CP3','CP4','CP5','CP6','CP7','CP8'])
data2
```

```python
n=len(data2)
for i in range(n):
    print("CP{:.0f} = {:.4f}Pregnancies + {:.4f}Glucose + {:.4f}BloodPressure + {:.4f}SkinThickness + \
{:.4f}Insu1in + {:.4f}BMI + {:.4f}DiabetesPedigreeFunction + {:.4f}Age".
format(i+1,data2['Pregnancies'][i],data2['Glucose'][i],data2['BloodPressure'][i], \
       data2['SkinThickness'][i],data2['Insulin'][i],data2['BMI'][i], \
       data2['DiabetesPedigreeFunction'][i],data2['Age'][i]))
```

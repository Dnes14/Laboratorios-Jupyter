## a. Obtenga y grafique la funci贸n de distribuci贸n ().


```python
#Librerias necesarias:
import pandas as  pd
import numpy as np
import math
import sympy as sp 

from sympy import integrate


import matplotlib.pyplot as plt
plt.style.use('classic')
```

```python
# Definimos las variables
x = np.array([4,5,6,7])
n = len(x)
N= 7
print("X=",x)
```

```python
# Hallamos la probabilidades
fx = np.zeros(n)
for i in range(n):
  k = x[i]
  fx[i] = math.comb(k-1, 3) / math.comb(7, 4)
print("f(x) =", fx)
```

```python
# Tabla
tabla1 =pd.DataFrame({"x":x,"f(x)=P(X=x)":fx})
tabla1
```

```python
# Grafico de f(x)

plt.vlines(x,0,fx,colors="blue", lw=5)
plt.ylabel("x")
plt.ylabel("f(x)")
plt.title("Grafico de f(x)")
```


## b. Calcule la esperanza, varianza y desviaci贸n est谩ndar para la variable aleatoria  redondeados con 3 decimales

```python
# Esperanza de X

Ex = np.sum(x*fx)
Ex=round(Ex,3)
print(Ex)
```

```python
# Esperanza de X^2

Ex2 = np.sum(x**2*fx)
Ex2=round(Ex2,3)
print(Ex2) 
```

```python
# varianza de X

Varx = Ex2 - (Ex)**2
Varx=round(Varx,3)
print(Varx)
```

```python

DesvX = math.sqrt(Varx)
DesvX=round(DesvX,3)
print(DesvX)
```

```python
fx1 = c*y1
print("f(x) = ", fx1,",0<=x<=1")
fx2 = c*y2
print("f(x) = ", fx2,",1<=x<=5")
```

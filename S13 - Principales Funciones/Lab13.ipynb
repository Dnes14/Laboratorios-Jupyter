```python
# Importamos las librerias

from scipy import stats

import numpy as np
import matplotlib.pyplot as plt 
plt.style.use('seaborn-v0_8-ticks')

# Configuraci√≥n warnings
import warnings
warnings.filterwarnings('ignore')
```


## Pregunta 1 :

Sea ùëã una variable aleatoria, tal que ùëã~ùëÉ(3). En un notebook de Jupyter y utilizando la librer√≠a ‚Äòstats‚Äô de ‚ÄòScipy‚Äô, haga lo siguiente:

### a. Genere un array para ùë• con valores de 0 a 9 y otro que contenga los valores de ùëì(ùë•).



```python
# X ~ P(3)

n = 3

fd = stats.poisson(n)
x = np.linspace(0, 9,10)

print("x =",x)
fd_x = fd.pmf(x)
print("f(x) =",fd_x)
```

### b. Halle la media y la varianza para la variable ùëã.



```python
# Media y varianza de x

media, varianza = fd.stats(moments='mv')
print("Media =",media)
print("Varianza =",varianza)
```

### c. Utilizando la librer√≠a ‚Äòmatplotlib.pyplot‚Äô, grafique la funci√≥n de distribuci√≥n ùëì(ùë•).

```python
# Gr√°fico de f(x)
plt.vlines(x,0,fd_x,colors="blue",lw=5,
         label='$\mu$ = {:3.1f}\n$\sigma^2$ = {:3.1f}'.format(float(media),float(varianza)))
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend(facecolor="orange",edgecolor="red",title="Descriptivos",loc=0)
plt.show
```

## Pregunta 2: Sea ùëã una variable aleatoria, tal que ùëã~ùëÅ(0,1). En un notebook de Jupyter y utilizando la librer√≠a ‚Äòstats‚Äô de ‚ÄòScipy‚Äô, haga lo siguiente:

### a.  Genere un array para ùë• con 100 valores que vayan de ‚Äì4 a 4 y otro que contenga los valores de ùëì(ùë•).


```python
#  X ~ N(m,var)

a= 0
b= 1

fd = stats.norm(a,b)
x = np.linspace(-4,4,100)
print("x =",x)
fd_x = fd.pdf(x) # funci√≥n de densidad evaluada en x
print("f(x) =",fd_x)
fd_rvs = fd.rvs(1000) #
```

### b. Halle la media y la varianza para la variable ùëã.


```python
# Media y varianza de x

media, varianza = fd.stats(moments='mv')
print("Media =",media)
print("Varianza =",varianza)
```


### c. Utilizando la librer√≠a ‚Äòmatplotlib.pyplot‚Äô, grafique la funci√≥n de distribuci√≥n ùëì(ùë•) que incluya tambi√©n un histograma con 500 valores muestreados con distribuci√≥n ùëã~ùëÅ(0,1)

```python
# Gr√°fico de f(x)
plt.plot(x,fd_x,lw=5,color="red",
           label='$\mu$ = {:3.1f}\n$\sigma^2$ = {:3.2f}'.format(float(media),float(varianza)))
plt.hist(fd_rvs,density=True)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend(facecolor="skyblue",edgecolor="purple",title="Descriptivos",loc=0)
plt.show
```

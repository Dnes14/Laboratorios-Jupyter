```python
# Importamos las librerias necesarias 
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline

# ================================
import warnings
warnings.filterwarnings('ignore')
```

--- 
## Parte A

```python
# Leemos los datos

data = pd.read_csv("vinos.csv", sep=";")
data
```

```python
# verificamos los tipos de datos
data.dtypes
```

```python
# Elimminamos 'quality'

data1 = data.drop("quality", axis = 1)
data1
```

```python
# Estandarizamos los datos
data_std = StandardScaler().fit_transform(data1)
A = np.cov(data_std, rowvar = False)
A
```

```python
# Hallamos los autovalores y autovectores
M = np.linalg.eig(A)
Autovalores, Autovectores = M[0], M[1]
print("Autovalores: \n",Autovalores)
print("\nAutovectores: \n",Autovectores)
```

--- 
## Parte B

```python
# Entrenando y escalando los datos
pca_pipe = make_pipeline(StandardScaler(),PCA())
pca_pipe.fit(data1)
modelo_pca = pca_pipe.named_steps['pca']
modelo_pca
```

```python
# Analizamos los componente convirtiendo en dataframe
df1 = pd.DataFrame(data=modelo_pca.explained_variance_,columns=['Autovalor'],
                  index=["λ1","λ2","λ3","λ4","λ5","λ6","λ7","λ8","λ9","λ10","λ11"])
df2 = pd.DataFrame(data=modelo_pca.explained_variance_ratio_*100,columns=['%Varianza'],
                  index=["λ1","λ2","λ3","λ4","λ5","λ6","λ7","λ8","λ9","λ10","λ11"])
df3 = pd.DataFrame(data=modelo_pca.explained_variance_ratio_.cumsum()*100,columns=['%Varianza acum.'],
                  index=["λ1","λ2","λ3","λ4","λ5","λ6","λ7","λ8","λ9","λ10","λ11"])
pd.concat([df1,df2,df3],axis=1)
```

```python
# convirtiendo el array a dataframe
data2 = pd.DataFrame(data= modelo_pca.components_,columns=data1.columns,
                    index = ['CP1','CP2','CP3','CP4','CP5','CP6','CP7','CP8','CP9','CP10','CP11'])
data2
```

--- 
## Parte C

Al analizar el porcentaje de varianza acumulada, se eligen 7 componentes principales, ya que estos tiene el 90.831906% de la varianza total. cumpliendo con la condicion de que al menos sean 90%.

```python
# Reprentamos el dataframe con la cantidad de componente elegidas
data3 = data2[:7]
data3
```

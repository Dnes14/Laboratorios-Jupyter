```python
# Importamos las librerias

from scipy import stats

import numpy as np
import matplotlib.pyplot as plt 
plt.style.use( 'seaborn-v0_8')
```

```python
# X ~ B(4,0.5)

n = 4
p = 0.5
```

```python
# Hamllamos f(x)

fd = stats.binom(n, p)
x = np.arange(n+1)

print("x =",x)
fd_x = fd.pmf(x)
print("f(x) =",fd_x)

```

```python
# Media y varianza de x

media, varianza = fd.stats(moments='mv')
print("Media =",media)
print("Varianza =",varianza)
```

```python
# Grafico de f(x)

plt.stem(x,fd_x)
plt.title("Grafico de f(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

```python
# Gráfico de f(x)
plt.vlines(x,0,fd_x,colors="blue",lw=5,
         label='$\mu$ = {:3.1f}\n$\sigma^2$ = {:3.1f}'.format(float(media),float(varianza)))
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend(facecolor="orange",edgecolor="red",title="Descriptivos",loc=0)
plt.show
```

```python
#  X ~ U[0,1]

a= 0
b= 1
```

```python
fd = stats.uniform(a,b)
x = np.linspace(-1,2,200)
print("x =",x)
fd_x = fd.pdf(x) # función de densidad evaluada en x
print("f(x) =",fd_x)
fd_rvs = fd.rvs(1000) # 1000 valores muestrados de la función de densidad
```

```python
# Media y varianza de x

media, varianza = fd.stats(moments='mv')
print("Media =",media)
print("Varianza =",varianza)
```

```python
# grafico de f(x)

plt.plot(x,fd_x)
plt.title("Grafico de f(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

```python
# Gráfico de f(x)
plt.plot(x,fd_x,lw=5,color="red",
           label='$\mu$ = {:3.1f}\n$\sigma^2$ = {:3.2f}'.format(float(media),float(varianza)))
plt.hist(fd_rvs,density=True)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend(facecolor="skyblue",edgecolor="purple",title="Descriptivos",loc=0)
plt.show
```

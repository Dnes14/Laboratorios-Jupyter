```python
# Importamos las librerias

import pandas as pd 
import numpy as np
from scipy import stats 
import random
```

```python
# Hallando f(x) y F(x)

x = np.array([100,150,200,250,300])
print("X =", x)
 
fx =np.array([0.30,0.20,0.30,0.15,0.05]) # funcion de distribucion evaluada en x
print("fx =", fx)

# funcion acumulada (propia)
fx_acumulada = np.cumsum(fx)
print("fx acumulada =", fx_acumulada)
```

```python
n = len(x)
Fx = np.zeros(n)
Fx[0] = fx[0]
for i in range(n):
    Fx[i] = Fx[i-1] + fx[i]
print("Fx =", Fx)
```

```python
tabla1 = pd.DataFrame({"x":x,"f(x)=P(X=x)":fx,"F(X)=P(X<=x)":Fx})
tabla1
```

```python
# geneamos 1000 numeros aleatorios
m= 1000
u= np.zeros(m)

for i in range(m):
    u[i] = random.uniform(0,1)

print(u)
```

```python
LC =200 # cantidad de licencias a comprar

LV = np.zeros(m)+100
for j in range(len(u)):
    for i in range(n):
        if (u[j] > Fx[i-1]) and (u[j] <= Fx[i]):
            LV[j] = min(x[i],LC)

print("Licencias vendidas:",LV)
```

```python
LD = LC-LV # diferencia de licencias
print("Licencias devueltas:",LD)
```

```python
CL = 75 # costo de licencia
ILV = 100 # ingreso por licencia vendida
ILD = 25 #Ingreso por licencia devuelta
CT= CL*LC # costo total
IV = ILV*LV # ingresos por licencias vendidas
ID = ILD*LD # ingresos por licencias devueltas
B = IV+ID-CT # beneficios
print("beneficios:",B)
BM = np.mean(B)
print("beneficios medios:",BM)
BD=np.std(B)
print("desviacion estandar del beneficio:",BD)
```

```python
#X1 N(20,3.4) distribucion del tiempo del primer servidor
u1 =20
Q1 = 3.4
# X2 N(22,2.6) distribucion del tiempo del segundo servidor
u2 = 22
Q2 = 2.6
```

```python
m =1000 # cantidad de simulaciones
u1 = np.zeros(m)
u2 = np.zeros(m)
for i in range(m):
    u1[i] = random.uniform(0,1)
u1
for i in range(m):
    u2[i] = random.uniform(0,1)
u2
```

```python
t1 = stats.norm(u1,Q1)
ts1 = t1.ppf(u1) #distribucion normal inversa
print("Tiempo del servidor1 =",ts1)
```

```python
t2 = stats.norm(u2, Q2)
ts2 = t2.ppf(u2) #distribucion normal inversa
print("Tiempo del servidor2 =",ts2)
```

```python
total = np.zeros(m)
for i in range(m):
    total[i]=max(ts1[i],ts2[i])

print("Tiempo total =",total) #tiempo maximo
```

```python
SMR = np.zeros(m)
for i in range(m):
    if(ts1[i]<ts2[i]):
        SMR[i]=1
    else:
        SMR[i]=2
print("Servidor mas rapido =",SMR)
```

```python
#generamos un data frame conlas frecuencias del servidor mas rapido

import collections 

contar = collections.Counter(SMR)

print(contar)

servidor = np.unique(SMR, return_counts=True)[0]
frecuencias = np.unique(SMR, return_counts=True)[1]
suma = sum(frecuencias)
pocentaje = frecuencias/suma*100

tabla2 = pd.DataFrame({'servidor mas rapido': servidor, 'frecuencia': frecuencias, 'pocentaje': pocentaje})
tabla2
```

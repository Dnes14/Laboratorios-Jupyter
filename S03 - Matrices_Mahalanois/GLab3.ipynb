```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("iris.csv")
data
```

```python
v_numericas = data.select_dtypes(["float64","int64"])
v_numericas = list(v_numericas.columns.values)
v_numericas
```

```python
x = data[v_numericas]
x
```

```python
#convirtiendo el dataframe a matriz

X = x.to_numpy()
X
```

```python
# Dimension de un matri

X.shape
```

```python
# Vector de medias para cada variables

Medias = np.mean(X,axis = 0)
Medias
```

```python
# Resta de la matriz de datos y el vector de medias

Resta= X - Medias
Resta
```

```python
# Matriz de covariana de las variables
Covarianza = np.cov(X, rowvar=False)
Covarianza
```

```python
# Inversa de la matriz  de covarianza de las variables

Inversa = np.linalg.inv(Covarianza)
Inversa
```

```python
Determinante = np.linalg.det(Covarianza)
Determinante
```

```python
Transpuesta = Resta[0].T
Transpuesta
```
